```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('data/pivot_data.csv')
df=df[['date','TSP_1', 'TSP_2', 'TSP_3', 'TSP_4', 'TSP_5', 'TSP_6', 'PM2.5_1',
       'PM2.5_2', 'PM2.5_3', 'PM2.5_4', 'PM2.5_5', 'PM2.5_6', 'O3_1', 'O3_2',
       'O3_3', 'O3_4', 'O3_5', 'O3_6', 'CO_1', 'CO_2', 'CO_3', 'CO_4', 'CO_5',
       'CO_6', 'NO2_1', 'NO2_2', 'NO2_3', 'NO2_4', 'NO2_5', 'NO2_6', 'SO2_1',
       'SO2_2', 'SO2_3', 'SO2_4', 'SO2_5', 'SO2_6']]
```

```python
df['date']=pd.to_datetime(df['date'], dayfirst=True)
```

```python
df.set_index('date', inplace=True)
```

```python
df.describe().transpose()
```

```python
pol=['TSP_1', 'TSP_2', 'TSP_3', 'TSP_4', 'TSP_5', 'TSP_6']
pol_label='TSP'
sns.set(style='whitegrid')
facecolor = '#eaeaf2'
fig, ax = plt.subplots(figsize=(8, 4), facecolor=facecolor)

ax = sns.boxplot(data=df[pol], 
                 palette='Set3', 
                 linewidth=1.2, 
                 fliersize=2, 
                 order=pol,
                 flierprops=dict(marker='o', markersize=4))

font_color = '#525252'
csfont = {'fontname':'Georgia'}
hfont = {'fontname':'Calibri'}

ax.set_ylabel('Concentrations of '+str(pol_label)+ ' (μg/$m^3$)', fontsize=12, color=font_color, **hfont)
ax.set_xlabel('Air Quality Stations', fontsize=12, color=font_color, **hfont)
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set(fontsize=12, color=font_color, **hfont)
    
title = str(pol_label)+' Concentrations across the HCMC'
#plt.title(title, fontsize=18, pad=10, color=font_color, **hfont)
```
